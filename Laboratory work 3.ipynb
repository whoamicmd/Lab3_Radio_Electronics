Розрахуємо струми та напруги для ланцюга (рис. 3.4) та побудуємо векторні діаграми для даних випадків. 

<center><img src="images/Elektric_scheme_3.png" width="450"/> <h4>Рисунок 3.4 - Електрична схема для розрахунку</h4></center>

Запишемо рівняння для схеми:
$$
\dot{U}_{N'N} = \frac{\frac{E_A}{Z_A} + \frac{E_B}{Z_B} + \frac{E_C}{Z_C}}{\frac{1}{Z_A} + \frac{1}{Z_B} + \frac{1}{Z_C}}
$$
$$
I_{\phi A} = \frac{E_A - \dot{U}_{N'N}}{Z_A}
$$
$$
I_{\phi B} = \frac{E_B - \dot{U}_{N'N}}{Z_B}
$$
$$
I_{\phi C} = \frac{E_C - \dot{U}_{N'N}}{Z_C}
$$

```python
import numpy as np

Ea = 120 * np.exp(1j * np.deg2rad(0))
Eb = 120 * np.exp(1j * np.deg2rad(-120))
Ec = 120 * np.exp(1j * np.deg2rad(+120))

R2 = 2000
R3 = 2000
C2 = 2.2e-6
C3 = 2.2e-6
f = 50
w = 2*np.pi*f

# Ємнісні опори
Xc2 = -1j / (w * C2)
Xc3 = -1j / (w * C3)

Za = 1e15 + 0j
Zb = R2 + Xc2
Zc = R3 + Xc3

# Напруга нейтралі
U_nn = (Ea/Za + Eb/Zb + Ec/Zc) / (1/Za + 1/Zb + 1/Zc)

print(f"|U_nn| = {abs(U_nn):.1f} В, angle = {np.angle(U_nn,deg=True):.1f}°")

# Струми
Ib = (Eb - U_nn) / Zb
Ic = (Ec - U_nn) / Zc

print(f"|Ib| = {abs(Ib):.4f} A, angle = {np.angle(Ib,deg=True):.2f}°")
print(f"|Ic| = {abs(Ic):.4f} A, angle = {np.angle(Ic,deg=True):.2f}°")

# Фазні напруги відносно нейтралі
Va = Ea - U_nn
Vb = Eb - U_nn
Vc = Ec - U_nn

print(f"|Va| =, {abs(Va):.4f}, V, angle, {np.angle(Va,deg=True):.4f}")
print(f"|Vb| =, {abs(Vb):.4f}, V, angle, {np.angle(Vb,deg=True):.4f}")
print(f"|Vc| =, {abs(Vc):.4f}, V, angle, {np.angle(Vc,deg=True):.4f}")
```
