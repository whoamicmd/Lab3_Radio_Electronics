Запишемо рівняння для схеми в комплексній формі:
$$
I_1 \cdot (I_1 + I_2 + I_3) = E1
$$
$$
I_2 \cdot R2 \cdot \frac{1}{j\omega C_2} + (I_1 + I_2 + I_3) = E2
$$
$$
I_3 \cdot R3 \cdot \frac{1}{j\omega C_3} + (I_1 + I_2 + I_3) = E3
$$

```python
import sympy as smp
import numpy as nmp
import math
from sympy import lambdify

E1_s = 0;
E2_s = 120;
E3_s = 120;
R2_s = 2000;
R3_s = 2000;
C1_s = 2.2 * 10 **-6;
C2_s = 2.2 * 10 **-6;
f_s = 50;
j = smp.I

# Оголошуємо символьні змінні для формули
I1, I2, I3, E1, E2, E3, R2, R3, C2, C3, W, f, = smp.symbols('I1 I2 I3 E1 E2 E3 R2 R3 C2 C3 \omega f')

eq = [-E2+I2*R2*(1/(j*W*C2)) + (I1+I2+I3), -E3+I3*R3*(1/(j*W*C1)) + (I1+I2+I3)]

sol = smp.solve(eq, [I2, I3])

V_b = R2 + 1 / (j * W * C2)
V_d = R3 + 1 / (j * W * C3)


V_b_f = lambdify([E1, E2, E3, R2, R3, C1, C2, W], V_b, 'numpy')
V_c_f = lambdify([E1, E2, E3, R2, R3, C1, C2, W], V_b, 'numpy')
print(nmp.abs(V_b_f(E1_s, E2_s, E3_s, R2_s, R3_s, C1_s, C2_s, 2*nmp.pi*f_s)))
print(nmp.abs(V_c_f(E1_s, E2_s, E3_s, R2_s, R3_s, C1_s, C2_s, 2*nmp.pi*f_s)))
```
